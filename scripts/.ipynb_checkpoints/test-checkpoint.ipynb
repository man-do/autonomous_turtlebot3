```python
def distance_2_points(point1, point2):
    # point is array of w,h coordinates of an image
    delta_w = point1[0]-point2[0]
    delta_h = point1[1]-point2[1]
    return math.sqrt(math.pow(delta_w, 2) + math.pow(delta_h, 2))
```

```python
ground_board_path = '/home/maverick/ros_ws/src/autonomous_turtlebot3/calib/chessboard_images/board_on_ground1.png'
img = cv2.cvtColor(cv2.imread(ground_board_path), cv2.COLOR_BGR2RGB)

img_gray = cv2.cvtColor(cv2.imread(ground_board_path), cv2.COLOR_RGB2GRAY)
mtx = pd.read_csv('/home/maverick/ros_ws/src/autonomous_turtlebot3/calib/mtx.csv', sep=' ', header=None)
dist = pd.read_csv('/home/maverick/ros_ws/src/autonomous_turtlebot3/calib/dist.csv', sep=' ', header=None)
newcameramtx = pd.read_csv('/home/maverick/ros_ws/src/autonomous_turtlebot3/calib/newcameramtx.csv', sep=' ', header=None)
mtx = np.asarray(mtx)
dist = np.asarray(dist)
newcameramtx = np.asarray(newcameramtx)

img = cv2.undistort(img,  mtx, dist, None, newcameramtx)
lwr = np.array([0, 0, 0])
upr = np.array([255, 30, 30])
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
msk = cv2.inRange(hsv, lwr, upr)
msk = cv2.bitwise_not(msk)

ret, corners = cv2.findChessboardCorners(msk, (3,3), None)

#img = cv2.drawChessboardCorners(msk, (3,3), corners,ret)

#img = cv2.circle(img, tuple(corners[0][0]), 5, (250,0,255), 1 )

pts1_index = [0, 2, 6, 8]
pts1 = corners[pts1_index]
pts2 = np.arange(8, dtype='float32').reshape(4, 1, 2)

pts2[0][0][0] = pts1[0][0][0]
pts2[0][0][1] = pts1[0][0][1]+240

pts2[1][0][0] = pts1[1][0][0]
pts2[1][0][1] = pts1[1][0][1] +240

pts2[2][0][0] = pts1[0][0][0]
pts2[2][0][1] = pts1[0][0][1]- \
                distance_2_points(pts1[0][0], pts1[2][0])+240

pts2[3][0][0] = pts1[1][0][0]
pts2[3][0][1] = pts1[1][0][1] - \
                distance_2_points(pts1[1][0], pts1[3][0])+240

w,h, ch = img.shape

plt.imshow(img)
```

```python
h_mtx = cv2.getPerspectiveTransform(pts1, pts2)
warped = cv2.warpPerspective(img, h_mtx, (w*3,h))

plt.imshow(warped)
```
